# Задание 1

Импортировать данные о ежемесячных ценах на нефть марки Brent по ссылке https://datahub.io/core/oil-prices/r/brent-month.csv Подсчитать средние цены на нефть, минимальное и максимальное значение, среднеквадратическое отклонение в течение 5 лет после месяца вашего рождения. Отобразить данные последнего года на графике.

```python
df = pd.read_csv('brent-month_csv.csv')
display(df.head())
```

Дата рождения: 2002-07

```python
df = df[(df['Date'] >= '2002-08-01') & (df['Date'] <= '2007-08-01')]
display(df)
```

```python
mean_number = df['Price'].mean()
print(f"Среднее значение: {mean_number:.2f}")
```

```python
max_price = df['Price'].max()
print(f"Максимальное значение: {max_price}")
```

```python
min_price = df['Price'].min()
print(f"Минимальное значение: {min_price}")
```

```python
std_number = df['Price'].std()
print(f"Среднеквадратическое отклонение: {std_number:.2f}")
```

```python
df = df[(df['Date'] >= '2006-08-01') & (df['Date'] <= '2007-08-01')]
fig = plt.figure(figsize=(14, 4))
plt.plot(df['Date'],df['Price'])
plt.show()
```

Импортировать данные о ценах на 4 криптовалюты https://www.kaggle.com/datasets/himanshunakrani/cryptocurrencies-dataset (DataExplorer, скачать csv-файл за ноябрь).
```python
df = pd.read_csv('crypto_data_updated_13_november.csv')
df.head()
```
